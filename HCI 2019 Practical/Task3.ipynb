```python
# Task 3.1

class Student:
    
    def __init__(self, StudID, Name, StudType, Result=None):
        self.StudID = StudID
        self.Name = Name
        self.StudType = StudType
        self.NoOfSub = '0'
        self.SkillArea = "Diploma"
        self.Result = Result
        

        
class Diploma(Student):
    
    def __init__(self, StudID, Name, StudType, NoOfSub, Result=None):
        super(Diploma, self).__init__(StudID, Name, StudType, Result)
        self.NoOfSub = NoOfSub
        

        
class SkillsCert(Student):
    
    def __init__(self, StudID, Name, StudType, SkillArea, Result=None):
        super(SkillsCert, self).__init__(StudID, Name, StudType, Result)
        self.SkillArea = SkillArea
```

```python
# Task 3.3

def read_file(filename):
    data = []
    with open(filename, 'r') as file:
        lines = file.readlines()
        for line in lines:
            temp = line.strip().split("|")
            
            if temp[2] == 'D':
                data.append(Diploma(temp[0], temp[1], temp[2], temp[4], temp[5]))

            else:
                data.append(SkillsCert(temp[0], temp[1], temp[2], temp[3], temp[5]))

    return data

students = read_file("STUDENT.txt")    
years = []
no_results = []

for student in students:
    if student.Result == '':
        if student.StudID[0:4] in years:
            no_results[years.index(student.StudID[0:4])].append(student)
            
        else:
            years.append(student.StudID[0:4])
            no_results.append([student])
            
no_results = [y for x, y in sorted(zip(years, no_results))]
years.sort()

n = len(years)
for i in range(n):
    print(f"Year: {years[i]}")
    print("-" * 50)
    
    for student in no_results[i]:
        diplomas = 0
        skills = 0
        if student.StudType == 'D': # diploma
            print("{0:6s} {1:12s} {2:4s} {3:4s}".format(student.StudID, student.Name, student.StudType, student.NoOfSub))
            diplomas += 1
            
        else: # skills certificate
            print("{0:6s} {1:12s} {2:4s} {3:12s}".format(student.StudID, student.Name, student.StudType, student.SkillArea))
            skills += 1
            
    print("{0:10s}{1}".format("Diplomas:", diplomas))
    print("{0:10s}{1}".format("Skills:", skills))
    print()
```
